```python
# Citric acid, chlorides, density, pH, sulphates, alcohol, alcohol_log1p, sulphates_log1p are the best categories/
# features. We will keep these and drop the rest.
df2 = data.drop(["volatile acidity", "residual sugar", "free sulfur dioxide", "total sulfur dioxide",
        "quality", "ResidualSugar_log1p", "TotalSulfurDioxide_log1p", "Chlorides_log1p"], axis=1)
```

```python
df2
```

```python
# split the data
df3 = data['quality']
df3
```

```python
# split the data
X_train, X_val, y_train, y_val = train_test_split(df2, df3, test_size =0.3, random_state=42)
```

```python
print("No. of samples in training set: ", X_train.shape[0])
print("No. of samples in validation set:", X_val.shape[0])
```

```python
print('Number of levels of quality in the training set:')
print(y_train.value_counts())

print('\n')
print('Number of levels of quality in the validation set:')
print(y_val.value_counts())
```

```python
type(y_val)
```

```python
test_y = y_val.astype("category")
type(test_y)
```

```python
test_y
```

```python
train_y = y_train.astype("category")
train_y
```

```python
model = LogisticRegression()
```

```python
test_y.value_counts()
```

```python
# confusion matrix

# classes determined by levels in training and validation sets
# quality levels were [3,4,5,6,7,8]
classes = ['3', '4', '5', '6', '7', '8']
cm = ConfusionMatrix(model, labels=np.unique(classes), percent=False)
# fits the passed model
cm.fit(X_train, train_y)
# Score runs predict() on test data and creates the confusion_matrix from scikit-learn.
cm.score(X_val, test_y)
for label in cm.ax.texts:
    label.set_size(20)
cm.poof()
```

```python
# Precision, Recall, and F1 Score
# set the size of the figure and the font size 
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 7)
plt.rcParams['font.size'] = 20
```

```python
visualizer = ClassificationReport(model, labels=np.unique(classes))
visualizer.fit(X_train, y_train)  # Fit the train data to visualizer
visualizer.score(X_val, y_val)  # Evaluate the model on the test data
g = visualizer.poof()
```

```python
# It seems like the highest scores among the three quality levels were found in quality level 5, and the lowest
# scores were found in quality level 7. Overall, not great scores though.
```

```python
# ROC and AUC
visualizer2 = ROCAUC(model)
visualizer2.fit(X_train, y_train)  # Fit the train data to visualizer
visualizer2.score(X_val, y_val)  # Evaluate the model on the test data
g = visualizer2.poof()
```
