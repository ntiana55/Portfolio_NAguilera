```python
import pandas as pd

df1 = pd.read_csv('/Users/Nicole/Documents/540DataPreparation/TermProject/Crime_Data_2010_2017.csv')
df1.head()
```

```python
# 1. missing values

df1.isnull()
```

```python
for c in df1.columns:
    miss = df1[c].isnull().sum()
    if miss>0:
        print("{} has {} missing value(s)".format(c,miss))
    else:
        print("{} has no missing values".format(c))
```

```python
df1.fillna('FILL')
```

```python
df2 = df1.dropna(axis=0)
df2
```

```python
# went from 1584316 to 13 rows, but they are all complete with no missing values
```

```python
# 2. outliers

df2.plot.box()
```

```python
# no outliers/not reasonable to measure by boxplot because there are categorical variables
```

```python
# 3. duplicate values for df2

df2.duplicated()
```

```python
# 4. replace headers

df3 = df2.rename({'DR Number': 'Dept Report Number', 'MO Codes': 'Method of Operation Codes'}, axis=1)
df3.head()
```

```python
# 5. Casing

df4 = df3.apply(lambda x: x.astype(str).str.upper())
df4
```

```python
df4.dtypes
```

```python
df4['Dept Report Number'] = pd.to_numeric(df4['Dept Report Number'])
df4['Area ID'] = pd.to_numeric(df4['Area ID'])
df4['Reporting District'] = pd.to_numeric(df4['Reporting District'])
df4['Crime Code'] = pd.to_numeric(df4['Crime Code'])
df4['Victim Age'] = pd.to_numeric(df4['Victim Age'])
df4['Crime Code 1'] = pd.to_numeric(df4['Crime Code 1'])
df4['Crime Code 2'] = pd.to_numeric(df4['Crime Code 2'])
df4['Crime Code 3'] = pd.to_numeric(df4['Crime Code 3'])
df4['Crime Code 4'] = pd.to_numeric(df4['Crime Code 4'])
```

```python
# reverted necessary columns back to numerics

df4.dtypes
```
